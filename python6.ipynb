```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("diabetes2.csv.xls")
df1 = df[["Glucose", "Outcome"]]
df1.head()
```

```python
plt.xlabel("Glucose")
plt.ylabel("Outcome")
plt.scatter(df.Glucose, df.Outcome, marker = "+", color = "blue")
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df[["Glucose"]], df.Outcome, test_size = 0.1)
x_test
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train.values, y_train.values)
```

```python
model.predict(x_test.values)
```
